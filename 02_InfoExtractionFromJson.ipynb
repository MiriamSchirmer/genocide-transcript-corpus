```python
import pandas as pd
import json
import os
from os.path import join
```

### Set the paths for input (dataset_json_path) and output (dataset_excel_path)

```python
dataset_json_path = "" # Path to JSON file
dataset_excel_path = "" # Path to dataset
```

### Extract information from JSON and output into an Excel file

```python
dataset_json = json.load(open(dataset_json_path))

#=====================================================================================
dict_documents = {}
for doc in dataset_json['documents']:
    dict_documents[doc['_id']] = {}
    dict_documents[doc['_id']]["name"] = doc['name']
    dict_documents[doc['_id']]["court"] = doc['name'][0:doc['name'].find("_")]
    dict_documents[doc['_id']]["plainText"] = doc['plainText']
#print(dict_documents['639a477497ad59b4cfc56821'].keys())
for index, elem in enumerate(dict_documents.keys()):
    print(f"{index+1}) {elem}")
    
    
dict_types = {}
for _type in dataset_json['types']:
    dict_types[_type['_id']] = {}
    dict_types[_type['_id']]["name"] = _type['name']
    
#print(dict_documents)

#=====================================================================================
counter = 1
list_annotations = list()
for annotation in dataset_json['annotations']:
    new_register = list()
    
    # Text - plainText
    try:
        new_register.append(annotation['document'])
        new_register.append(annotation["_id"])
        plain_text = dict_documents[annotation['document']]["plainText"][annotation['start']:annotation['end']]
        plain_text = ' '.join(plain_text.splitlines())#.replace("\r", " ").replace('\n', ' ')
        new_register.append(plain_text)
    except KeyError as ex:
        #print(f"Document [{annotation['document']}] was not found")
        continue
    # Role
    new_register.append(dict_types[annotation['type']]["name"])
    # Trauma - default value = 0
    new_register.append(0)
    # Court
    new_register.append(dict_documents[annotation['document']]["court"])
    # Date
    # etc.
    
    ### Save new register to list of annotations
    list_annotations.append(new_register)

df = pd.DataFrame(list_annotations, columns=["id_document", "id_annotation", "text", "role", "trauma", "court"])
df.to_excel(dataset_excel_path, index=False, encoding='latin-1')

#=====================================================================================
print("\n\n\t\tExtraction of information was successfully executed!!!")
```
